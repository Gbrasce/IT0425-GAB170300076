#Regresión Lineal

**La regresión lineal** es un algoritmo de aprendizaje automático supervisado, también se considera el tipo más simple de algoritmo predictivo de aprendizaje automático.

* Hay una suposición básica que hacemos para que funcione la regresión lineal.
* Se supone que la relación entre la variable independiente (X) y la variable objetivo (Y) es de naturaleza líneal.
* De lo contrario, el modelo dará resultados irrelevantes.
* La palabra líneal sugiere que la función utilizada para la predicción es una función lineal.
* Esta función se puede representar como:
$$y = mx + b$$

* Esta es la bien conocida ecuación de una línea recta.
* En términos de regresión lineal, $y$ es la variable dependiente, representa el valor conocido.
* $x$ es la variable independiente.
* $m$ es la pendiente de la recta.
* $b$ es la intercepción en el eje de la variable dependiente $y$.
* $m$ y $b$ son los coeficientes que necesitamos optimizar para ajustar la línea de regresión a nuestros datos (puntos).

##Cálculo de los coeficientes

Para calcular los coeficientes $m$ y $b$ necesitamos la fórmula de la varianza y covarianza.

$$ m = \text{cov}(x, y) \; / \; \text{var}(x) $$
$$ b = \bar{y} \; —  \; m * \bar{x} $$

donde
* $\bar{y}$ es la media de $y$
* $\bar{x}$ es la media de $x$

la varianza se calcula con la siguiente fórmula:
$$\text{var}(x, y) = \frac{\sum (x_i - \bar{x})^2}{n} $$

y la covarianza se calcula con la siguiente fórmula:
$$\text{cov}(x, y) = \frac{\sum (x_i - \bar{x}) \; (y_i - \bar{y})}{n}$$



```python
# número de coordenadas (puntos)
# tamaño del vector
n = np.size(x)
print(n)
```

```python
# media de las coordenadas X
media_x = np.mean(x)
print(media_x)
```

```python
# media de las coordenadas Y
media_y = np.mean(y)
print(media_y)
```

```python
# grafíca los puntos usando X, Y
plt.scatter(x, y, marker='o', c='r', ec='k')
# grafíca el punto que representa la media
plt.scatter(media_x, media_y, marker='D', c='y', ec='k', s=200)
# etiquetas en ejes
plt.xlabel('x')
plt.ylabel('y');
```

```python
# suma del cuadrado de los errores
# Sum of Squared Errors (SSE)
SSE_xy = np.sum(y*x) - (n * media_y * media_x)
SSE_xx = np.sum(x*x) - (n * media_x * media_x)
```

```python
# grafíca los puntos usando X, Y
plt.scatter(x, y, marker='o', c='r', ec='k')
# grafíca el punto que representa SSE
plt.scatter(SSE_xy, SSE_xx, marker='p', c='c', ec='k', s=200)
# etiquetas en ejes
plt.xlabel('x')
plt.ylabel('y');
```

```python
# calcula los coeficientes de regresión
m = SSE_xy / SSE_xx
b = media_y - m*media_x
```

```python
# utilizando la ecuación de la recta
# se obtiene la respuesta estimada
y_estimada = m*x + b
print(y_estimada)
```

```python
# grafíca los puntos
plt.scatter(x, y, marker='o', c='y', ec='k');

# grafíca la línea de regresión
plt.plot(x, y_estimada, color = "b")

# se agregan etiquetas en los ejes
plt.xlabel('x')
plt.ylabel('y')

# función para mostrar la gráfica
plt.show()
```

## Usando una función
Función para calcular los coeficientes de regresión.


```python
# dadas las coordenadas X,Y
# se estima los coeficientes

def calculo_coef(x, y):
  # número de coordenadas (puntos)
  n = np.size(x)

  # media de las coordenadas X,Y
  media_x = np.mean(x)
  media_y = np.mean(y)

  # Sum of Squared Errors (SSE)
  # calcula la desviación cruzada y la desviación sobre x
  SSE_xy = np.sum(y*x) - (n * media_y * media_x)
  SSE_xx = np.sum(x*x) - (n * media_x * media_x)

  # calcula los coeficientes de regresión
  β = SSE_xy / SSE_xx
  α = media_y - β*media_x

  return α, β
```

```python
# coeficientes
α, β = calculo_coef(x, y)

# estimación de y
y_estimada = α + β*x

# gráfica los puntos
plt.scatter(x, y, marker='D', c='y', ec='k');

# gráfica la línea de regresión
plt.plot(x, y_estimada, color = "r")

# agrega etiquetas
plt.xlabel('x')
plt.ylabel('y')

# muestra todo
plt.show()
```

## Predicción (prueba)

```python
# nuevas coordenadas en X
xt = np.array([10, 11, 12, 13])

# estimación de y
y_prediccion = α + β*xt
```

Las nuevas coordenadas en Y, son calculadas utilizando el modelo líneal.

```python
# gráfica los puntos
plt.scatter(x, y, marker='D', c='y', ec='k')
plt.scatter(xt, y_prediccion, marker='o', c='g', ec='k');

# gráfica la línea de regresión
plt.plot(x, y_estimada, color = "r")
plt.plot(xt, y_prediccion, "--", color = "g")

# agrega etiquetas
plt.xlabel('x')
plt.ylabel('y')

# muestra todo
plt.show()
```
